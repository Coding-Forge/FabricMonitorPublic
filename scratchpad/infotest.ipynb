```python
workspaceScanResults = guid_list
```

```python
groups_of_500 = [workspaceScanResults[i:i+500] for i in range(0, len(workspaceScanResults), 500)]

# Split each group into 16 subgroups as evenly as possible
subgroups = []
for group in groups_of_500:
    items_per_subgroup = len(group) // 16
    remainder = len(group) % 16
    start_index = 0
    subgroup = []
    for i in range(16):
        end_index = start_index + items_per_subgroup
        if i < remainder:
            end_index += 1
        subgroup.append(group[start_index:end_index])
        start_index = end_index
    subgroups.append(subgroup)
```
